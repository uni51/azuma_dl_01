## 4.5 活性化関数

### 4.5.1 ステップ関数

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return np.where(x<=0, 0, 1)

x = np.linspace(-5, 5)
y = step_function(x)

plt.plot(x, y)
plt.show()
```

### 4.5.2 シグモイド関数

```python
%matplotlib inline

import numpy as np
import matplotlib.pylab as plt

def sigmoid_function(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-5, 5)
y = sigmoid_function(x)

plt.plot(x, y)
plt.show()
```

### 4.5.3 tanh

```python
%matplotlib inline

import numpy as np
import matplotlib.pylab as plt

def tanh_function(x):
    return np.tanh(x)

x = np.linspace(-5, 5)
y = tanh_function(x)

plt.plot(x, y)
plt.show()
```

### 4.5.4 ReLU

```python
%matplotlib inline

import numpy as np
import matplotlib.pylab as plt

def relu_function(x):
    return np.where(x <= 0, 0, x)

x = np.linspace(-5, 5)
y = relu_function(x)

plt.plot(x, y)
plt.show()
```

### 4.5.5 Leaky ReLU

```python
%matplotlib inline

import numpy as np
import matplotlib.pylab as plt

def leaky_relu_function(x):
    return np.where(x <= 0, 0.01*x, x)

x = np.linspace(-5, 5)
y = leaky_relu_function(x)

plt.plot(x, y)
plt.show()
```

### 4.5.6 恒等関数

```python
%matplotlib inline

import numpy as np
import matplotlib.pylab as plt

x = np.linspace(-5, 5)
y = x

plt.plot(x, y)
plt.show()
```

### 4.5.7 ソフトマックス関数

```python
import numpy as np

def softmax_function(x):
    return np.exp(x)/np.sum(np.exp(x)) # ソフトマックス関数

y = softmax_function(np.array([1,2,3]))
print(y)
```
